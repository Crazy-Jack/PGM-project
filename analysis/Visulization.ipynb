# PACS data (P + A to C)

### Load data

```python
baseline_path = "/home/tianqinl/Code/PGM-project/train_related/shuffle_domain_pacs_ctest/shuffle_weight_1.000000_learningrate_0.001000_startsepoch_100_isshuffle_0_domains_2"

loss_p = np.load(baseline_path + "/loss_source_1_.npy")
loss_a = np.load(baseline_path + "/loss_source_2_.npy")
train_acc_p = np.load(baseline_path + "/source_acc_1_.npy")
train_acc_a = np.load(baseline_path + "/source_acc_2_.npy")
test_acc_p = np.load(baseline_path + "/source_test_acc_1_.npy")
test_acc_a = np.load(baseline_path + "/source_test_acc_2_.npy")
test_acc_c = np.load(baseline_path + "/target_test_acc_.npy")
```

```python
shuffle_domain_path = "/home/tianqinl/Code/PGM-project/train_related/shuffle_domain_pacs_ctest/shuffle_weight_1.000000_learningrate_0.001000_startsepoch_100_isshuffle_1_domains_2"
sloss_p = np.load(shuffle_domain_path + "/loss_source_1_.npy")
sloss_a = np.load(shuffle_domain_path + "/loss_source_2_.npy")
strain_acc_p = np.load(shuffle_domain_path + "/source_acc_1_.npy")
strain_acc_a = np.load(shuffle_domain_path + "/source_acc_2_.npy")
stest_acc_p = np.load(shuffle_domain_path + "/source_test_acc_1_.npy")
stest_acc_a = np.load(shuffle_domain_path + "/source_test_acc_2_.npy")
stest_acc_c = np.load(shuffle_domain_path + "/target_test_acc_.npy")
```

### Loss (Baseline)

```python
plt.plot(loss_p)
plt.plot(loss_a)
plt.legend(["Training Loss - Photos", "Training Loss - Art"])
plt.title("Training Loss (Baseline P + A -> C)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```

### Loss (Shuffled)

```python
# Loss
plt.plot(sloss_p)
plt.plot(sloss_a)
plt.legend(["Training Loss - Photos", "Training Loss - Art"])
plt.title("Training Loss (Domain Shuffled P + A -> C)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```

### Accuracy (Baseline)

```python
plt.plot(train_acc_p)
plt.plot(train_acc_a)
plt.plot(test_acc_p)
plt.plot(test_acc_a)
plt.plot(test_acc_c)
plt.legend(["train_p", "train_a", "test_p", "test_a", "test_c"])
plt.title("Classification Accuracy (Baseline P + A -> C)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```

### Accuracy (Shuffled)

```python
plt.plot(strain_acc_p)
plt.plot(strain_acc_a)
plt.plot(stest_acc_p)
plt.plot(stest_acc_a)
plt.plot(stest_acc_c)
plt.legend(["train_p", "train_a", "test_p", "test_a", "test_c"])
plt.title("Classification Accuracy (Domain Shuffled P + A -> C)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```

### Pairwise comparsion

```python
# P
plt.plot(strain_acc_p)
plt.plot(train_acc_p)
plt.plot(stest_acc_p)
plt.plot(test_acc_p)
plt.legend(["Domain Shuffled - train", "baseline - train", "Domain Shuffled - test", "baseline - test"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Classification Accuracy Comparison on P")
```

```python
# A
plt.plot(strain_acc_a)
plt.plot(train_acc_a)
plt.plot(stest_acc_a)
plt.plot(test_acc_a)
plt.legend(["Domain Shuffled - train", "baseline - train", "Domain Shuffled - test", "baseline - test"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Classification Accuracy Comparison on A")
```

```python
# C
plt.plot(stest_acc_c)
plt.plot(test_acc_c)
plt.legend(["Domain Shuffled - test", "baseline - test"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Classification Accuracy Comparison on C")
```
